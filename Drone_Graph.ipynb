```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
df = pd.read_csv('record_weight_0.6_0.4.csv')

# 그래프를 2x1 서브플롯으로 생성
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# Selection Method별로 다른 색상과 마커 사용
colors = ['blue', 'red', 'green']
markers = ['o', 's', '^']

# Distance 그래프
for method in [0, 1, 2]:
    method_data = df[df['Selection Method'] == method]
    ax1.plot(method_data[' Number of Destinations'],
             method_data[' Distance'],
             color=colors[method],
             marker=markers[method],
             label=f'Method {method}')

ax1.set_xlabel('Number of Destinations')
ax1.set_ylabel('Distance')
ax1.set_title('Distance by Number of Destinations for each Selection Method')
ax1.grid(True)
ax1.legend()

# Energy 그래프
for method in [0, 1, 2]:
    method_data = df[df['Selection Method'] == method]
    ax2.plot(method_data[' Number of Destinations'],
             method_data[' Energy'],
             color=colors[method],
             marker=markers[method],
             label=f'Method {method}')

ax2.set_xlabel('Number of Destinations')
ax2.set_ylabel('Energy')
ax2.set_title('Energy by Number of Destinations for each Selection Method')
ax2.grid(True)
ax2.legend()

# 서브플롯 간 간격 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```
